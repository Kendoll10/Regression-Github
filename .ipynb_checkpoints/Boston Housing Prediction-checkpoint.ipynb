# Using The Random Sample Consensus (RANSAC) Algorithm

```python
X = df["RM"].values.reshape(-1, 1)
X.shape
```

```python
y = df["MEDV"].values
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
model = RANSACRegressor()
```

```python
model.fit(X, y)
```

```python
inlier_mask = model.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y = model.predict(line_X.reshape(-1, 1))
```

```python
sns.set(context="notebook", style="darkgrid", palette="deep", font="sans-serif", font_scale=1)
plt.figure(figsize=(12,12))
plt.scatter(X[inlier_mask], y[inlier_mask], c="blue", marker="o", label="inliers")
plt.scatter(X[outlier_mask], y[outlier_mask], c="brown", marker="s", label="outliers")
plt.plot(line_X, line_y, color="red")
plt.xlabel("average number of rooms per dwelling")
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc="upper right")
plt.show()
```

```python
model.estimator_.coef_
```

```python
model.estimator_.intercept_
```

# Performance evaluation of Regression model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df["LSTAT"].values.reshape(-1, 1)
```

```python
y = df["MEDV"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
Ir = LinearRegression()
```

```python
Ir.fit(X_train, y_train)
```

```python
y_train_predict = Ir.predict(X_train)
```

```python
y_test_predict = Ir.predict(X_test)
```

# method 1 : Residual analysis

```python
plt.figure(figsize=(12, 12))
plt.scatter(y_train_predict, (y_train_predict-y_train), c="blue", marker="o", label="Training Data")
plt.scatter(y_test_predict, (y_test_predict-y_test), c="brown", marker="*", label="Test Data")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.legend(loc="upper right")
plt.hlines(y=0, xmin=-10, xmax=50, linestyles="solid", color="k")
plt.xlim=([-10, 50])
plt.show()
```

# Method 2: Mean Square Error (MSE)

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_predict)
```

```python
mean_squared_error(y_test, y_test_predict)
```

# Method 3: Coefficient of Determination R^2

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_predict)
```

```python
r2_score(y_test, y_test_predict)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(X_train.reshape(-1, 1), y_train)

y_train_predict = model.predict(X_train.reshape(-1, 1))
y_test_predict = model.predict(X_test.reshape(-1, 1))
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(y_train_predict, y_train_predict-y_train, c="blue", marker="o", label="Training Data")
plt.scatter(y_test_predict, y_test_predict-y_test, c="brown", marker="*", label="Test Data")
plt.xlabel("Predicted value")
plt.ylabel("Residual")
plt.legend(loc="upper right")
plt.hlines(y=0, xmin=-3, xmax=33, color="k")
plt.xlim=([-5, 35])
plt.show()
```

# 1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
pipeline.fit(X, y)
```

```python
result = pd.DataFrame(list(zip(pipeline.steps[1][1].coef_, df.columns)), columns=["coefficient", "name"]).set_index("name")
np.abs(result).sort_values(by="coefficient", ascending=False)
```
