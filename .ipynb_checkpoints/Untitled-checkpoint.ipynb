```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
theta = 3
alpha = 0.1
dat = []

for i in range(0, 15):
    res = alpha * 2 * theta
    print("{0:.4f} {1:4f}".format(theta, res))
    dat.append([theta, theta ** 2])
    theta = theta - res
```

```python
tmp = pd.DataFrame(dat)
tmp
```

```python
plt.figure(figsize=(12, 12))
plt.plot(np.linspace(-2, 4, 100), np.linspace(-2, 4, 100) ** 2)
plt.scatter(tmp.iloc[:,0], tmp.iloc[:,1], marker="*")
plt.xlabel("theta")
plt.ylabel("J(theta)")
plt.show()
```

```python
theta = 3
alpha = 0.01
dat = []

for i in range(0, 10):
    res = alpha * (4 * theta ** 3 + 2 * theta)
    print("{0:.4f} {1:.4f}".format(theta, res))
    dat.append([theta, (theta ** 4 + theta ** 2)])
    theta = theta - res
```

```python
tmp = pd.DataFrame(dat)
tmp
```

```python
plt.figure(figsize=(12, 12))
x_grid = np.linspace(-2, 4, 100)
plt.plot(x_grid, (x_grid ** 4 + x_grid ** 2))
plt.scatter(tmp.iloc[:,0], tmp.iloc[:,1], marker="*")
plt.xlabel("theta")
plt.ylabel("J(theta)")
plt.show()
```

```python
from sklearn.datasets import load_boston
```

```python
boston_data = load_boston()
```

```python
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
```

```python
df.head()
```

```python
X = df[["LSTAT"]]
y = boston_data.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y.reshape(-1,1)).flatten()
```

```python
alpha = 0.0001
w_ = np.zeros(1 + X_std.shape[1])
cost_ = []
n_ = 100

for i in range(n_):
    y_pred = np.dot(X_std, w_[1:]) + w_[0]
    errors = (y_std - y_pred)
    
    w_[1:] += alpha * X_std.T.dot(errors)
    w_[0] += alpha * errors.sum()
    
    cost = (errors ** 2).sum() / 2
    cost_.append(cost)
```

```python
plt.figure(figsize=(12, 12))
plt.plot(range(1, n_ + 1), cost_)
plt.xlabel("Epoch")
plt.ylabel("SSE")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# X is the 10x10 Hilbert matrix
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis])
y = np.ones(10)

# ###########################################################################
# Compute paths

n_alphas = 200
alphas = np.logspace(-10, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)

# ###########################################################################
# Display results

plt.figure(figsize=(10,8))
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = 1. / (np.arange(1, 11) + np.arange(0, 10)[:, np.newaxis])
y = np.ones(10)
```

```python
n_alphas = 200
alphas = np.logspace(-10, -2, n_alphas)

coefs = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)
```

```python
plt.figure(figsize=(10,10))
ax = plt.gca()

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])
plt.xlabel("alpha")
plt.ylabel("weight")
plt.title("Ridge coefficients as a function of the regularization")
plt.axis('tight')
plt.show()
```

```python
%matplotlib inline
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
n_samples = 100
rng = np.random.randn(n_samples) * 10
y_gen = 0.5 * rng + 2 * np.random.randn(n_samples)

lr = LinearRegression()
lr.fit(rng.reshape(-1, 1), y_gen)
model_predict = lr.predict(rng.reshape(-1,1))

plt.figure(figsize=(12, 12));
plt.scatter(rng, y_gen);
plt.plot(rng, model_predict);
print("coefficient: ", lr.coef_)
```

```python
idx = rng.argmin()
y_gen[idx]=-200
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(rng, y_gen)

o_lr = LinearRegression(normalize=True)
o_lr.fit(rng.reshape(-1, 1), y_gen)
o_model_predict = o_lr.predict(rng.reshape(-1,1))

plt.plot(rng, o_model_predict)
print("coefficient ", o_lr.coef_)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 1, normalize=True)
ridge_model.fit(rng.reshape(-1, 1), y_gen)
ridge_predict = ridge_model.predict(rng.reshape(-1, 1))

plt.figure(figsize=(12, 12))
plt.scatter(rng, y_gen)
plt.plot(rng, ridge_predict)
print("coefficient ", ridge_model.coef_)
plt.show()
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=0.4, normalize=True)
lasso_model.fit(rng.reshape(-1, 1), y_gen)
lasso_predict = lasso_model.predict(rng.reshape(-1, 1))

plt.figure(figsize=(12,12))
plt.scatter(rng, y_gen)
plt.plot(rng, lasso_predict)
print("coefficient ", lasso_model.coef_)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_model = ElasticNet(alpha=0.02, normalize=True)
elastic_model.fit(rng.reshape(-1, 1), y_gen)
elastic_predict = elastic_model.predict(rng.reshape(-1, 1))

plt.figure(figsize=(12, 12))
plt.scatter(rng, y_gen)
plt.plot(rng, elastic_predict)
print("coefficient ", elastic_model.coef_)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
np.random.seed(42)
n_samples = 100

X = np.linspace(0, 10, 100)
rng = np.random.randn(n_samples) * 100

y = X ** 3 + rng + 100

plt.figure(figsize=(12, 12))
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X.reshape(-1, 1), y)
model_predict = lr.predict(X.reshape(-1,1))

plt.figure(figsize=(12, 12));
plt.scatter(X, y);
plt.plot(X, model_predict);
print(r2_score(y, model_predict))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(-1, 1))
```

```python
lr_2 = LinearRegression()
lr_2.fit(X_poly, y.reshape(-1, 1))
y_predict = lr_2.predict(X_poly)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(X, y)
plt.plot(X, y_predict)
print(r2_score(y, y_predict))
```
