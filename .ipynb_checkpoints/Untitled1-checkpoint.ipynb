```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(1000)
y = 3 * x + np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
lr = model.fit(X_train.reshape(-1,1), y_train)
```

```python
y_train_predict = model.predict(X_train.reshape(-1,1))
```

```python
y_test_predict = model.predict(X_test.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_predict, y_train_predict-y_train, c="blue", marker="s", label="training data")
plt.scatter(y_test_predict, y_test_predict-y_test, c="orange", marker="*", label="test data")
plt.xlabel("Predicted value")
plt.ylabel("Residual")
plt.hlines(y=0, xmin=-3, xmax=33, lw=2, colors="k")
plt.xmin=([-3, 33])
plt.legend(loc="upper right")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_predict)
```

```python
mean_squared_error(y_test, y_test_predict)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_predict)
```

```python
r2_score(y_test, y_test_predict)
```

# multiple regression with statsmodels

```python
from sklearn.datasets import load_boston
```

```python
boston_loads = load_boston()
```

```python
df = pd.DataFrame(boston_loads.data, columns=boston_loads.feature_names)
```

```python
df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X = df
```

```python
y = boston_loads.target
```

```python
y
```

```python
X_constant = sm.add_constant(X)
```

```python
pd.DataFrame(X_constant)
```

```python
model = sm.OLS(y, X_constant)
```

```python
lr = model.fit()
```

```python
lr.summary()
```

```python
form1 = smf.ols(formula="y ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT", data=df)
lr = form1.fit()
lr.summary()
```
